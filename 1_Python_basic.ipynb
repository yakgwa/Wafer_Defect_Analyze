##연습문제)반복문

예제) for문 학급평균점수
* A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.
* A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
* for문을 사용하여 A 학급의 평균 점수를 구해 보자.
* 힌트1) '평균 = 전체합 / A학급의 학생의 수' 계산식에서  전체합은 for문으로 계산한다. 
* 힌트2) 'A학급의 학생의 수' len()함수 사용
* 결과값) 79

## 예제 정답코드

```python
for x in range(1,31) :
  print(x)
```

```python
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
add = 0
for a in A:
    # add = add + a   # A학급의 점수를 모두 더한다.
    add +=  a

# add
average = add / len(A) # 평균을 구하기 위해 총 점수를 총 학생수로 나눈다.
print(average) # 평균 79.0이 출력된다.
```

```python
data_A = {'ID': [1,2,3],
          'name': ['Kim', 'Choi', 'Lee']}
df_A = pd.DataFrame(data = data_A, columns = ['ID', 'name'])

data_B = {'ID': [2,3,4], 
          'age': [33, 15, 20]}
df_B = pd.DataFrame(data = data_B, columns = ['ID', 'age'])
```

pd.merge(데이터프레임1, 데이터프레임2, on = , how =)
   *      on = 'Key', how= 'inner', 'outer', 'left', 'right'

```python
df_inner = pd.merge(df_A, df_B, on='ID', how='inner')
df_inner
```

```python
df_outer = pd.merge(df_A, df_B, on='ID', how='outer')
df_outer
```

```python
df_left = pd.merge(df_A, df_B, on='ID', how='left')
df_left
```

```python
df_right = pd.merge(df_A, df_B, on='ID', how='right')
df_right
```
